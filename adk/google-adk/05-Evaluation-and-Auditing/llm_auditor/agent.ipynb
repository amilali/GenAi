LLM Auditor for verifying & refining LLM-generated answers using the web.
<br>
from google.adk.agents import SequentialAgent<br>
from .sub_agents.critic import critic_agent<br>
from .sub_agents.reviser import reviser_agent<br>
import logging<br>
import google.cloud.logging<br>
logging.basicConfig()<br>
cloud_logging_client = google.cloud.logging.Client()<br>
cloud_logging_client.setup_logging()<br>
llm_auditor = SequentialAgent(<br>
    name='llm_auditor',<br>
    description=(<br>
        'Evaluates LLM-generated answers, verifies actual accuracy using the'<br>
        ' web, and refines the response to ensure alignment with real-world'<br>
        ' knowledge.'<br>
    ),<br>
    sub_agents=[critic_agent, reviser_agent],<br>
)<br>
root_agent = llm_auditor
